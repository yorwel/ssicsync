```python
import os
import pandas as pd

# Path to the folder containing the PDF files
folder_path = "AR Report Data/"

# Get a list of PDF files in the folder
pdf_files = [file for file in os.listdir(folder_path) if file.endswith('.pdf')]

# Create a DataFrame with the file names
coy_df = pd.DataFrame({'File Name': pdf_files})

# Extracting only the name part into a new column
coy_df['Coy Name'] = coy_df['File Name'].apply(lambda x: ' '.join(x.split('.')[:-1])[:-5].strip() if '.' in x else x[:-9].strip())
# coy_df

# Convert 'Coy Name' column values to a list
company_queries = coy_df['Coy Name'].tolist()
company_queries
```

```python
# 3rd Layer, Webscrap Company Main Webpage

import pandas as pd
from bs4 import BeautifulSoup

# Replace 'path_to_file' with the actual path to your CSV file
file_path = "Webscrape from Linkedin.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Function to extract HTML content for each URL
def extract_html(url):
    try:
        driver.get(url)
        return driver.page_source
    except Exception as e:
        print("Error:", e)
        html_content = ""

# Create a new column to store HTML content
df['Org Website Content'] = None

# Initialize Selenium WebDriver (make sure you have installed the appropriate webdriver for your browser)
# driver = webdriver.Chrome()  # Example: Chrome WebDriver

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    org_website = row['Org Website']
    if pd.notnull(org_website):  # Check if Org Website is not NaN
        print(f"Extracting HTML for URL: {org_website}")
        html_content = extract_html(org_website)
        if html_content:
            # Store HTML content in the new column
            df.at[index, 'Org Website Content'] = html_content
        else:
            print("Failed to fetch HTML content for URL:", org_website)
    else:
        print("Org Website is NaN for row:", index)


# Scan through Main page html and extract linked (href tags) with keywords

def extract_and_filter_links(html_content, filter_strings):
    href_urls = []
    if html_content:
        soup = BeautifulSoup(html_content, 'html.parser')
        # Find all anchor tags (a) and extract the href attribute
        for a_tag in soup.find_all('a', href=True):
            href_urls.append(a_tag['href'])
        # Filter out links containing any of the strings in the filter_strings list
        filtered_links = [link for link in href_urls if any(string in link.lower() for string in filter_strings)]
        return filtered_links
    else:
        return []

# Define the list of keywords to filer -> for about, company profile etc
filter_strings = ['about', 'profile']

# Apply the combined function to the 'Org Website Content' column and store the results in a new column
df['Filtered Links'] = df['Org Website Content'].apply(lambda x: extract_and_filter_links(x, filter_strings))

# Remove duplicates in the list in 'Filtered Links' column
df['Filtered Links'] = df['Filtered Links'].apply(lambda x: list(set(x)))


### Transformation
# Iterate over each row
for index, row in df.iterrows():
    org_website = row['Org Website']  # Get the 'Org Website' string for the current row
    filtered_links = row['Filtered Links']  # Get the list of strings from the 'Filtered Links' column
    
    # Iterate over each string in the list
    for i in range(len(filtered_links)):
        if not filtered_links[i].startswith('http'):  # Check if it's a suffix (doesn't start with 'http')
            if not filtered_links[i].startswith('/'):  # Check if it doesn't start with '/'
                filtered_links[i] = '/' + filtered_links[i]  # Append '/' at the start
            filtered_links[i] = org_website + filtered_links[i]  # Append the suffix to the end of the 'Org Website' string

    # Update the 'Filtered Links' column with the modified list
    df.at[index, 'Filtered Links'] = filtered_links

df
```

```python
# Cleanup

from urllib.request import urlopen
from bs4 import BeautifulSoup

# Data Cleanup

def extract_text_from_html(html_content):
    # If the HTML content is not empty
    if html_content:
        # Parse the HTML content
        soup = BeautifulSoup(html_content, features="html.parser")
        
        # Remove script and style elements
        for script in soup(["script", "style"]):
            script.extract()  # Remove the element
        
        # Get text content
        text = soup.get_text(separator=' ')
        
        # Define strings to remove from the text
        strings_to_remove = ['\n', '>', '|']  # Add strings you want to remove here
        
        # Remove specified strings from the text
        for string_to_remove in strings_to_remove:
            text = text.replace(string_to_remove, '')
        
        return text.strip()  # Return the stripped text
    else:
        return None  # Return None if the HTML content is empty


# Apply the function to the 'Org Website Content' column and store the results in a new column
df1['Extracted Text'] = df1['Org Website Content'].apply(extract_text_from_html)

df1
```

```python
# Open pkl file

import pickle
import pandas as pd

# Open and read the pickle file
with open('pkl/extractedWebsite.pkl', 'rb') as file1:
    data1 = pickle.load(file1)

# Convert to DataFrame if not already a DataFrame
extractedWebsite = pd.DataFrame(data1)

# Stack the DataFrame to combine all values into a single column
stacked_series = extractedWebsite.stack().reset_index()

# Rename the columns for clarity
stacked_series.columns = ['index', 'pdf_name', 'url']

# Define a function to append "https://" to URLs
def add_https(url):
    if url.startswith('https://'):
        return url
    elif url.startswith('http://'):
        return url.replace('http://', 'https://')
    else:
        return 'https://' + url

# Apply the function to add "https://" to URLs
stacked_series['url'] = stacked_series['url'].apply(add_https)

# Drop the index column
combined_df = stacked_series.drop('index', axis=1).sort_values(by='pdf_name').drop_duplicates().reset_index(drop=True)

pd.set_option('display.max_rows', None)
combined_df
```
